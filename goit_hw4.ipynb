```python
# Перетворюємо стовпець WorkExp на числовий тип (ігноруємо помилки)
df_public['WorkExp'] = pd.to_numeric(df_public['WorkExp'], errors='coerce')

# Розрахунок мір центральної тенденції
mean_exp = df_public['WorkExp'].mean()
median_exp = df_public['WorkExp'].median()
mode_exp = df_public['WorkExp'].mode()[0] if not df_public['WorkExp'].mode().empty else None

print(f"Середнє значення досвіду роботи: {mean_exp:.2f}")
print(f"Медіана досвіду роботи: {median_exp}")
print(f"Мода досвіду роботи: {mode_exp}")
```

```python
# які унікальні значення має стовпець RemoteWork
print(df_public['RemoteWork'].unique())
```

```python
# скільки респондентів працюють віддалено
remote = df_public[df_public['RemoteWork'] == 'Remote']
print("Кількість респондентів, які працюють віддалено:", remote.shape[0])
```

```python
# Прибираємо рядки, де немає даних у стовпці LanguageHaveWorkedWith,
# і перевіряємо, чи містить рядок слово "Python"
python_users = df_public['LanguageHaveWorkedWith'].dropna().apply(lambda x: 'Python' in x)

# Обчислюємо відсоток респондентів, які зазначили Python серед мов, з якими працювали
percent_python = python_users.sum() / len(df_public) * 100

# Виводимо результат з округленням до двох знаків після коми
print(f"Відсоток респондентів, які програмують на Python: {percent_python:.2f}%")
```

```python
# Пошук кількості респондентів, які навчались через онлайн-курси або сертифікації
count_online = df_public['LearnCode'].dropna().str.contains('Online Courses or Certification', case=False).sum()

print("Кількість респондентів, які навчались через онлайн курси:", count_online)
```

```python
# True для рядків, де вказано Python (регістронезалежно), NaN -> False
mask_python = df_public['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False)

# респонденти, які працювали з Python
df_python = df_public[mask_python].copy()

# Переконаємось, що тип = число
df_python['ConvertedCompYearly'] = pd.to_numeric(df_python['ConvertedCompYearly'], errors='coerce')

# Групуємо по країні і рахуємо count, mean та median для компенсації (ігноруємо NaN у компенсації)
result = (df_python
          .groupby('Country')['ConvertedCompYearly']
          .agg(count='count', mean_compensation='mean', median_compensation='median')
          .sort_values(by='mean_compensation', ascending=False))

# Показуємо топ-10 країн (можеш змінити або вивести весь результат)
print(result.head(10))
```

```python
# вибір лише потрібних колонок
df_comp = df_public[['ConvertedCompYearly', 'EdLevel']].copy()

# Перетворення компенсації у числовий формат
df_comp['ConvertedCompYearly'] = pd.to_numeric(df_comp['ConvertedCompYearly'], errors='coerce')

# Видалення рядків, де компенсація відсутня
df_comp = df_comp.dropna(subset=['ConvertedCompYearly'])

# сортування по спаданню
df_top5 = df_comp.sort_values(by='ConvertedCompYearly', ascending=False).head(5)

# Вивід рівня освіти топ-5 респондентів
print(df_top5[['ConvertedCompYearly', 'EdLevel']])
```

```python
# копія датафрейму з потрібними колонками
df_age_lang = df_public[['Age', 'LanguageHaveWorkedWith']].copy()

# видалення рядків зневказаним віком
df_age_lang = df_age_lang.dropna(subset=['Age'])

# Нова колонка 'UsesPython': True якщо містить 'Python', False якщо ні
df_age_lang['UsesPython'] = df_age_lang['LanguageHaveWorkedWith'].fillna('').str.contains('Python', case=False)

# Підрахунок загальної кількісті респондентів у кожній віковій категорії
total_by_age = df_age_lang.groupby('Age')['UsesPython'].count()

# Підрахунок кількості тих, хто використовує Python у кожній віковій категорії
python_by_age = df_age_lang.groupby('Age')['UsesPython'].sum()

# Обчислення відсотка
python_percent_by_age = (python_by_age / total_by_age * 100).round(2)

# результат у вигляді таблиці
print(python_percent_by_age.sort_index())
```

```python
# Видалення рядків, де відсутня компенсація або тип роботи
df_filtered = df_public.dropna(subset=['ConvertedCompYearly', 'RemoteWork', 'Industry'])

# значення 75-го перцентиля компенсації
percentile_75 = df_filtered['ConvertedCompYearly'].quantile(0.75)

# ті, хто має компенсацію >= 75-го перцентиля
#     і працює віддалено (у колонці RemoteWork є підрядок 'Remote')
df_top_remote = df_filtered[
    (df_filtered['ConvertedCompYearly'] >= percentile_75) &
    (df_filtered['RemoteWork'].str.contains('Remote', case=False, na=False))
]

# підрахунок кількості респондентів за індустріями
industry_counts = df_top_remote['Industry'].value_counts()

# Вивід ТОП-10 найпоширеніших індустрій
print("=== Найпоширеніші індустрії серед топ-25% за компенсацією і remote ===")
print(industry_counts.head(10))
```
